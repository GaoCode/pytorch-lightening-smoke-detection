```python
def generate_preds_dicts(image_preds_dir):
    image_preds = np.loadtxt(image_preds_dir, delimiter=',', dtype=str)

    fires = {}
    gt = []
    preds = []

    positive_preds_dict = {} # Holds positive predictions for each fire
    negative_preds_dict = {} # Holds negative predictions for each fire
    positive_times_dict = {} # Holds timestamps of positives for each fire

    for (image_name, image_pred, loss) in image_preds:
        fire_name = util_fns.get_fire_name(image_name)
        if fire_name in metadata['mislabeled_fires']: continue

        if image_pred == 'True' or image_pred == '1':
            image_pred = 1
        else:
            image_pred = 0

        if fire_name not in fires:
            fires[fire_name] = []
            positive_preds_dict[fire_name] = []
            negative_preds_dict[fire_name] = []
            positive_times_dict[fire_name] = []

        fires[fire_name].append(image_pred)
        gt.append(util_fns.get_ground_truth_label(image_name))
        preds.append(image_pred)

        if util_fns.get_ground_truth_label(image_name) == 0:
            negative_preds_dict[fire_name].append(image_pred)
        else:
            positive_preds_dict[fire_name].append(image_pred)
            positive_times_dict[fire_name].append(util_fns.image_name_to_time_int(image_name))

    return fires, gt, preds, positive_preds_dict, negative_preds_dict, positive_times_dict
```

```python
def recalculate_test_metrics(image_preds_dir):
    fires, gt, preds, positive_preds_dict, negative_preds_dict, positive_times_dict = generate_preds_dicts(image_preds_dir)

    acc = sklearn.metrics.accuracy_score(gt, preds)
    f1 = sklearn.metrics.f1_score(gt, preds)
    precision = sklearn.metrics.precision_score(gt, preds)
    recall = sklearn.metrics.recall_score(gt, preds)
    avg_ttd, med_ttd, std_ttd = util_fns.calculate_time_to_detection_stats(positive_preds_dict, positive_times_dict)

    data = [acc, f1, precision, recall, avg_ttd]

    acc = str(round(acc*100, 2))
    f1 = str(round(f1*100, 2))
    precision = str(round(precision*100, 2))
    recall = str(round(recall*100, 2))
    avg_ttd = str(round(avg_ttd, 2))

    print(acc+'/'+f1+'/'+precision+'/'+recall+'/'+avg_ttd)
    
    return data
```

# Fire-Wise Proportions

```python
def calculate_fire_statistics(image_preds_dir, threshold=1):
    """
    Graph 0: Proportion of Fires with >= X Correct Positives
    Graph 1: Proportion of Fires with >= X Correct Consecutive Positives
    Graph 2: Proportion of Fires with >= <threshold> Correct Positives X Mins after Start
    Graph 3: Proportion of Fires with >= <threshold> Correct Consecutive Positives X Mins after Start
    Graph 4: Proportion of Fires with >= X Consecutive False Positives
    Graph 5: Average # of Correct Positives X Mins after Start
    Graph 6: Average # of Correct Consecutive Positives X Mins after Start
    Graph 7: Accuracy at Time X
    Graph 8: Visualization of Accuracy per Fire
    """
    fires, gt, preds, positive_preds_dict, negative_preds_dict, positive_times_dict = generate_preds_dicts(image_preds_dir)

    graph0_count = []
    graph1_count = []
    graph2_count = {}
    graph3_count = {}
    graph4_count = []
    graph5_count = {}
    graph6_count = {}
    graph7_count = {}

    for fire in positive_preds_dict:
        max_consecutive = 0
        current_consecutive = 0

        graph0_count.append(sum(positive_preds_dict[fire]))

        for i, pred in enumerate(positive_preds_dict[fire]):
            if pred == 1:
                current_consecutive += 1
                if current_consecutive > max_consecutive:
                    max_consecutive = current_consecutive
            else:
                current_consecutive = 0

            if str(i) not in graph2_count:
                graph2_count[str(i)] = []
                graph3_count[str(i)] = []
                graph5_count[str(i)] = []
                graph6_count[str(i)] = []
                graph7_count[str(i)] = []
            graph2_count[str(i)].append(int(sum(positive_preds_dict[fire][:i+1]) >= threshold))
            graph3_count[str(i)].append(int(max_consecutive >= threshold))
            graph5_count[str(i)].append(sum(positive_preds_dict[fire][:i+1]))
            graph6_count[str(i)].append(max_consecutive)
            graph7_count[str(i)].append(pred)

        graph1_count.append(max_consecutive)

    graph0_count = np.array(graph0_count)
    graph1_count = np.array(graph1_count)

    graph0_return = []
    graph1_return = []
    for i in range(max(graph0_count)):
        graph0_return.append(len(graph0_count[graph0_count > i])/len(graph0_count))
        graph1_return.append(len(graph1_count[graph1_count > i])/len(graph1_count))

    graph2_return = []
    graph3_return = []
    graph5_return = []
    graph6_return = []
    graph7_return = []
    graph8_return = []
    graph9_return = []
    for i in range(len(graph2_count)):
        graph2_return.append(sum(graph2_count[str(i)]) / len(graph2_count[str(i)]))
        graph3_return.append(sum(graph3_count[str(i)]) / len(graph3_count[str(i)]))
        graph5_return.append(sum(graph5_count[str(i)]) / len(graph5_count[str(i)]))
        graph6_return.append(sum(graph6_count[str(i)]) / len(graph6_count[str(i)]))
        
        ones = np.ones(len(graph7_count[str(i)]))
        graph7_return.append(sklearn.metrics.accuracy_score(graph7_count[str(i)], ones))


    for fire in negative_preds_dict:
        max_consecutive = 0
        current_consecutive = 0

        for i, pred in enumerate(negative_preds_dict[fire]):
            if pred == 1:
                current_consecutive += 1
                if current_consecutive > max_consecutive:
                    max_consecutive = current_consecutive
            else:
                current_consecutive = 0

        graph4_count.append(max_consecutive)

    graph4_count = np.array(graph4_count)
    graph4_return = []
    for i in range(max(graph4_count)):
        graph4_return.append(len(graph4_count[graph4_count > i])/len(graph4_count))

    return np.array([np.array(graph0_return), 
                     np.array(graph1_return), 
                     np.array(graph2_return), 
                     np.array(graph3_return), 
                     np.array(graph4_return), 
                     np.array(graph5_return), 
                     np.array(graph6_return),
                     np.array(graph7_return)
                    ])
```

# Inference Time

```python
# Source: https://towardsdatascience.com/the-correct-way-to-measure-inference-time-of-deep-neural-networks-304a54e5187f
model_type_list = ["RawToTile_ResNet", "TileToTile_LSTM", "TileToTileImage_SpatialViT"]

model_list = nn.ModuleList()
for model_type in model_type_list:
    model_list.append(globals()[model_type](
                     tile_embedding_size=1000,
                     num_tiles=40,
                     num_tiles_height=5,
                     num_tiles_width=8,
                     series_length=3,
                     backbone_size='small'))

device = torch.device("cuda")
model_list.to(device)


dummy_input = torch.randn(1,40,3,3,224,224,dtype=torch.float16).to(device)
starter, ender = torch.cuda.Event(enable_timing=True), torch.cuda.Event(enable_timing=True)
repetitions = 300
timings=np.zeros((repetitions,1))

bbox_label = None
# bbox_label = [[{"boxes": torch.zeros((0, 4), dtype=torch.float32).to(device),
#                                   "labels": torch.zeros(0, dtype=torch.int64).to(device),
#                                   "area": torch.zeros(0, dtype=torch.float32).to(device),
#                "masks": torch.zeros((0, 1040, 1856), dtype=torch.uint8).to(device)}]]

# GPU-WARM-UP
for _ in range(10):
    outputs = None
    x = dummy_input
    for model in model_list:
        outputs, x = model(x, tile_outputs=outputs, bbox_labels=bbox_label)

# MEASURE PERFORMANCE
with torch.no_grad():
    for rep in range(repetitions):
        starter.record()
        outputs = None
        x = dummy_input
        for model in model_list:
            outputs, x = model(x, tile_outputs=outputs, bbox_labels=bbox_label)
        ender.record()
        
        # WAIT FOR GPU SYNC
        torch.cuda.synchronize()
        curr_time = starter.elapsed_time(ender)
        timings[rep] = curr_time

mean_syn = np.sum(timings) / repetitions
std_syn = np.std(timings)

print('Time Mean:', mean_syn / 1000)
print('Time SD:', std_syn / 1000)
```

```python
def calculate_pos_weight(labels_dict):
    tile_sums = []
    
    for image in labels_dict:
        x = np.array(labels_dict[image])
        tile_sums.extend(x)

    tile_sums = np.array(tile_sums)
    num_tiles = len(labels_dict[image])
    
    total_pos_tiles = len(tile_sums)
    total_neg_tiles = 144 * 40 * num_tiles 
    total_tiles = total_pos_tiles + total_neg_tiles

    smoke_threshold = 250
    target_tiles = len(tile_sums[tile_sums > smoke_threshold])

    neg_percent = (total_tiles - target_tiles) / total_tiles
    pos_weight = neg_percent / (1-neg_percent)
    
    return pos_weight
```
